```python
%matplotlib inline
from matplotlib import pyplot as pp
import numpy as np

import face_data
import eigenfaces
```

Loading all faces

```python
faces = face_data.load_data()
```

Calculating the eigenvectors over the covariance matrix built over 1000 faces

```python
w, v = eigenfaces.calc( faces[:1000] )
```

```python
pp.subplot(221)
pp.matshow(v[:,0].reshape((32,32)),fignum=False)
pp.title("Eigenvalue %.3f" % (w[0]),y=1.2)

pp.subplot(222)
pp.matshow(v[:,1].reshape((32,32)),fignum=False)
pp.title("Eigenvalue %.3f" % (w[1]),y=1.2)

pp.subplot(223)
pp.matshow(v[:,2].reshape((32,32)),fignum=False)
pp.title("Eigenvalue %.3f" % (w[2]),y=1.2)

pp.subplot(224)
pp.matshow(v[:,3].reshape((32,32)),fignum=False)
pp.title("Eigenvalue %.3f" % (w[3]),y=1.2)

pp.tight_layout()
```
